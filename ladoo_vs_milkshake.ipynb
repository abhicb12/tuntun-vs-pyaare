```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y') # Specifying the correct format '%d-%m-%Y' to handle the dates present in your file.
df['Date']
```

```python
df['Date'].dt.to_period('M')
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total_Sale_Value'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
```

```python
bins = [0, 12, 18, 30, 45, 60, 100]
labels = ['Kids', 'Teens', 'Young Adults', 'Adults', 'Middle Age', 'Senior']
df['Age_Group'] = pd.cut(df['Customer_Age'], bins=bins, labels=labels)

age_sales = df.groupby('Age_Group')['Total_Sale_Value'].sum()
age_sales.plot(kind='bar', title='Sales by Age Group')
```
